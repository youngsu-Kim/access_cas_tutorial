```python
# Parallely execution
# Run 4 degree 5 calculations
# based on https://doc.sagemath.org/html/en/reference/parallel/sage/parallel/decorate.html#sage.parallel.decorate.parallel

@parallel(p_iter='fork')
def worker(deg):
    # pari.allocatemem(2*1024*1024*1024)

    R.<x>=QQ[]
    f = R.random_element(degree=deg)
    if f.is_irreducible():
        return f.splitting_field('w')
    else:
        return "reducible"

list(worker([5,5,5]))
```
